```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
bank_notes = pd.read_csv("Banknote-authentication-dataset-.csv")
bank_notes.count()
```

```python
v1 = bank_notes["V1"]
v2 = bank_notes["V2"]
```

```python
# Getting the mean, min and max

v1_min = v1.min()
v1_max = v1.max()
v1_mean = v1.mean()

v2_min = v2.min()
v2_max = v2.max()
v2_mean = v2.mean()
```

```python
# Normalizing the datasets
for i in range(0, len(v1)):
    v1[i] = (v1[i] - v1_mean)/(v1_max - v1_min)
    
for i in range(0, len(v2)):
    v2[i] = (v2[i] - v2_mean)/(v2_max - v2_min)
```

```python
# Invoking the Sklearn Library
from sklearn.cluster import KMeans
km_result = KMeans(n_clusters=2).fit(np.column_stack((v1, v2)))
```

```python
v1_value_0 = []
v1_value_1 = []
v2_value_0 = []
v2_value_1 = []

for i in range(0, len(v1)):
    if (km_result.labels_[i] == 0):
        v1_value_0.append(v1[i])
        v2_value_0.append(v2[i])
    else:
        v1_value_1.append(v1[i])
        v2_value_1.append(v2[i])
```

```python
#Plotting the values and cluster centres
fig, ax = plt.subplots(figsize= (9,5))
ax.scatter(v1_value_0, v2_value_0, c= "red")
ax.scatter(v1_value_1, v2_value_1, c="blue")
ax.scatter(km_result.cluster_centers_[:,0], km_result.cluster_centers_[:,1], c="green" , s= 400, marker = "s")
ax.set_xlabel("V1-Axis")
ax.set_ylabel("V2-Axis")
ax.set_title("Plot of Genuine and Counterfeit banknotes")
plt.legend(["Counterfeit note","Genuine note",])
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(bank_notes[["V1", "V2"]])
    sse.append(km.inertia_)
```

```python
plt.xlabel("K")
plt.ylabel("SSE")
plt.plot(k_rng,sse)
```
